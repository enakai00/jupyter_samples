```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import math
import pickle
import matplotlib.animation as animation
```

```python
%matplotlib nbagg
images = []
with open('20160517_longrun/log_for_l2_05_0978', mode='rb') as f:
    best_log = pickle.load(f)

fig = plt.figure(figsize=[3,3])

for log in best_log:
    th = log[0]
    center = [50, 50]
    top = [50+50*np.sin(th),50-50*np.cos(th)]
    if log[1] > 0:
        color = 'red'
    else:
        color = 'blue'
    plt.ylim(0, 100)
    plt.xlim(0, 100)
    plt.xticks([])
    plt.yticks([])
    image = plt.plot([center[0], top[0]], [center[1], top[1]], color=color)
    images.append(image)

ani = animation.ArtistAnimation(fig, images,interval=15,repeat_delay=1000)
plt.show()
```

```python
%matplotlib inline
with open('20160517_longrun/log_for_l2_05_0978', mode='rb') as f:
    best_log = pickle.load(f)
            
ths = np.array(best_log)[:,0]
dths = ths[1:] - ths[:-1]

fig = plt.figure(figsize=(20,4))
subplot = fig.add_subplot(1,2,1)
subplot.set_ylim(-1,1)
subplot.set_xlim(0,300)
x = range(len(best_log))
subplot.plot(x,-np.cos(ths))

subplot = fig.add_subplot(1,2,1)
subplot.set_ylim(-1,1)
subplot.set_xlim(0,300)
x = range(len(best_log))
subplot.plot(x, 0.8*np.array(best_log)[:,1])

subplot = fig.add_subplot(1,2,1)
subplot.set_ylim(-1,1)
subplot.set_xlim(0,300)
x = range(len(dths))
subplot.plot(x, dths)
```
